##1. if/else
Sólo tiene dos posibles respuestas (V) o (F).
Sintaxis
```
if condicion):
  aqui se ejecuta código **si** la condicion es verdadera.
else:
  aqui se ejecuta el codigo **si** la condicion es falsa
  ```

````python
lluvia=input("Está lloviendo?")
````
